# SEABORN Tutorial

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

# Distribution Plots

- distplot
- joinplot
- pairplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Correlation with Heatmaps

A correlation heatmap used colored cells, typically in monochromatic scale, to show a 2D correletion matrix(table) between two discrete dimensions or event types. It is very important in Feature Selection.

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
```

# JoinPlot

A join plot allows you to study the  relationship between 2 numeric variables. The central chart display correlation. It is usually a scatterplot. a hexbin plot, a 2D histogram or a 2D density plot.

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind = 'hex'); # kind specifies the type of plot in the middle
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg'); # reg here refers to making a scatter plot with a regression line too, as well as a density lines on the axis plots
```

# Pair Plot

A "pairs plot" is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this showing all variables of paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='time')
```

```python
sns.pairplot(df, hue='smoker')
```

```python
sns.pairplot(df, hue='day')
```

# Dist plot

Dist Plot, or distribution plot, helps to check the distribution of the columns feature.

This is basically a histogram plot so we can just use that function.

```python
sns.distplot(df['tip'])
```

```python
# since .distplot() what depricated we have to use the .histplot() function

sns.histplot(df['tip'], kde=True, stat = 'density', linewidth=0)
```

```python
sns.histplot(df['tip'], kde=False)
```

# Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about:

- box plots
- violin plot
- count plot
- bar plot

```python
## Count Plots

sns.countplot(x = 'smoker', data = df)
```

```python
sns.countplot(x='day', data = df)
```

```python
sns.countplot(x='time', data = df)
```

```python
sns.countplot(x='day', data = df, hue = 'time')
```

```python
sns.countplot(y='sex', data = df)
```

```python
## Bar Plot

sns.barplot(x='total_bill', y='sex', data = df)
```

```python
sns.barplot(x='smoker', y='total_bill', data = df)
```

```python
df.head()
```

# Box Plot

A boxplot and whisker plot (sometimes called a boxplot) is a graph that presents information form a five-family summary.

```python
sns.boxplot(data = df, x = 'smoker', y = 'total_bill')
```

```python
sns.boxplot(data = df, x = 'sex', y = 'total_bill')
```

```python
sns.boxplot(data = df, x = 'day', y = 'total_bill', palette='rainbow')
```

```python
sns.boxplot(data = df, orient = 'v')
```

```python
# categorize data based on the different categories

sns.boxplot(x='total_bill', y = 'day', hue='smoker', data=df)
```

# Violin Plot

Violin plots help us to see both the distrubution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill', y = 'day', data = df, palette = 'rainbow')
```

```python
sns.violinplot(x='total_bill', y = 'day', data = df, hue = 'smoker')
```
