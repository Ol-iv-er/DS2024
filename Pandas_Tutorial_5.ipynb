# CSV

Understanding CSV data

```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
           'x,y,1\n'
           'a,b,c\n'
           'c,d,3')
```

```python
type(data)
```

```python
# StringIO(data) <- this makes a file in memory that then can be read in
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from specific columns

df2=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```python
df2.head()
```

```python
df2.to_csv('csv_test_file.csv')
```

```python
## Specifying columns data types

data = ('a,b,c,d\n'
           '1,2,3,4\n'
           '5,6,7,8\n'
           '9,10,11\n')
```

```python
df=pd.read_csv(StringIO(data), dtype = object) # we specified the data type to be an object, which in this case is string
```

```python
df
```

```python
df['a'][1]
```

```python
df.loc[0,'b'] 
```

```python
df=pd.read_csv(StringIO(data), dtype={'b':int, 'c':float, 'a':'Int64'}) # specifiy the data type for each column
```

```python
df
```

```python
df['a'][1] # now returns an integer
```

```python
## check the data type of each column
df.dtypes
```

```python
## Index columns and training delimiters
```

```python
data=('index,a,b,c\n'
         '4,apple,bat,5.7\n'
         '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data)) # the index values are not in the right place
```

```python
pd.read_csv(StringIO(data), index_col=0) # here we specify the index location, it can be any column
```

```python
data=('a,b,c\n'
         '4,apple,bat,\n'
         '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data)) # makes the integers the index as the parameter is set to that
```

```python
pd.read_csv(StringIO(data), index_col=False)
```

```python
## Combining usecols and index_col
data=('a,b,c\n'
         '4,apple,bat,\n'
         '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data), usecols=['b','c'], index_col=False)
```

```python
## Quoting and Escape Characters. Very useful in NLP

data = 'a,b,\n"Hello, \\"Bob\\", nice to see you", 5'
```

```python
pd.read_csv(StringIO(data), escapechar='\\')
```
