## Fidelity Stock Overview

```python
# pip install yfinance
# pip install pandas --upgrade
import yfinance as yf
import pandas as pd
```

```python
fxaix = yf.Ticker('FXAIX')

# get historical market data
hist_data = fxaix.history(period='max')
```

```python
hist_data.head(20)
```

```python
yearly_investment = 7000
```

```python
data = hist_data[(hist_data.index.month == 1)]
#data.index = pd.to_datetime(data.index)
```

```python
data
```

```python
lst = []
year = 2010
for i, row in data.iterrows():
    if i.year != year:
        lst.append(i.strftime("%Y-%m-%d"))
        year = i.year
```

```python
year_data = data.loc[lst, :]
```

```python
year_data['Shares_owned'] = yearly_investment / year_data.Close
```

```python
year_data
```

```python
todays_price = hist_data.tail(1)
todays_price['Close']
```

```python
total_shares = sum(year_data.Shares_owned) 
total_shares
```

```python
invested_amount = yearly_investment * len(year_data)
invested_amount
```

```python
stock_total_value = total_shares * todays_price
stock_total_value.loc[:,'Close']
```

```python
total_gains = stock_total_value - invested_amount
total_gains.loc[:,'Close']
```
