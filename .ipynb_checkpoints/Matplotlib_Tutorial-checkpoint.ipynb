# MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-orited API for embedding plots into applications using general-purpose GUI tookits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline
## '%matplotlib inline' command ensures that the plot of the sine wave is displayed directly below the code cell. ##
```

```python
import numpy as np
```

```python
# simple examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
## plotting using matplotlib

## plt scatter

plt.scatter(x,y, c='purple')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Graph in 2-d")
plt.savefig('test_plot.png')
# plt.show() ## do not this code since in jupyter notebook we have the "%matplotlib inline"
```

```python
## plotting using the plot function for more control

y = x*x

plt.plot(x, y,'r*', linestyle='dashed', linewidth=1.5, markersize=9)
plt.title('2D Diagram')
plt.xlabel('X axis')
plt.ylabel('Y axis')
## plt.plot(x, y, color='r--', linewidth=1.5, markersize=9) <- this is the same as before but using short hand
```

```python
## Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y, 'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'go--')

plt.subplot(2,2,3)
plt.plot(x,y,'b*-')

plt.subplot(2,2,4)
plt.plot(x,y,'purple')
```

```python
x = np.arange(0, 50)

y=3*x + 10

plt.title('Matplotlib demo')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.plot(x,y)
```

```python
# access inbulit values in numpy
np.pi
```

```python
# Compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)

plt.title('Sine Wave Form')

# Plot the points using matplotlib
plt.plot(x,y)
# plt.show()
```

```python
# Use subplots to make some more plots
# compute the x and y coordinares

x = np.arange(0, 2 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up the subplot grid that has height 2 and width 1
# and set the first subplot as active

plt.subplot(2,1,1)
plt.title('Sine Function')
plt.plot(x,y_sin)

plt.subplot(2,1,2)
plt.title('Cosine Funtion')
plt.plot(x, y_cos, 'r--')
```

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y) # used for bar plots
plt.bar(x2,y2, color = 'g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

# Histograms

```python
a = np.array([22,87,5,43,56,72,78,11,20,23,56,63,22,17,8])
plt.hist(a, bins = 12)
plt.title('Histogram') # bins is 10 by default
```

```python
a.mean()
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)] # this gets 100 steps of the intervals 0:1, 0:2, 0:4

# rectangular box plot
plt.boxplot(data, vert = True, patch_artist=True);
```

# Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java' # this declaration makes it into a tuple of strings
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True) # autopct is set to _._ format percentages (%)

plt.axis('equal');
```
